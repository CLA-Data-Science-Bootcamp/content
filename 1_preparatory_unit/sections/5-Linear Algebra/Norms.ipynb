#Definition
A norm is a function from the vector space to the nonnegative real numbers that measures the distance from the origin, on an intuitive level. In machine learning, we usually measure the size of vectors using a function called a norm.
##L1 Norm
The L1 norm of a vector is simply the sum of the absolute values of the vector.




```
import numpy as np
from numpy.linalg import norm
a = np.array([4, 2, 3])
l1 = norm(a, 1)
print(l1)
```

##L2 norm
The L2 norm that is the square root of the sum of the squared vector values.

```
a = np.array([4, 2, 3])
l1 = norm(a)
print(l1)
```
