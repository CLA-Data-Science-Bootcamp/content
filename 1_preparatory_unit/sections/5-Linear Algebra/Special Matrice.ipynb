##Triangular matrix 
A triangular matrix is a square matrix in which all the elements are null except either the ones above or under the main diagonal. A square matrix is called “lower triangular” if all the entries above the main diagonal are zero. Similarly, a square matrix is called upper triangular if all the entries below the main diagonal are zero.
